```python
# Create the calculator using the function  
# Hand Made By OP

def add(a,b):
    return a+b 

def sub(a,b):
    return a-b

def mul(a,b):
    return a*b

def div(a,b):
    return a/b

print("Simple Calculator")

a = float(input("Enter first number:"))
b = float(input("Enter second number:"))

print("1. Add")
print("2. Suntract")
print("3. Multiply")
print("4. Divide")

choice = int(input("Enter choice (1-4):"))

if choice ==1:
    print("Result:" , add(a,b))
elif choice ==2:
    print("Result:" , sub(a,b))
elif choice ==3:
    print("Result:" , mul(a,b))
elif choice ==4:
    print("Result:" , div(a,b))
else:
    print("Invalid choice")
```
