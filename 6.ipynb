```python
import matplotlib.pyplot as plt # Data visiulization 
```

```python
x = [1,2,3,4,5]
y = [2,3,4,6,10]

plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
data = [45,50,60,70,72,80,85,88,90,95,96,98]

plt.hist(data ,bins=5)
plt.title("Histrogram of scores")
plt.xlabel("Score")
plt.ylabel("frequency")
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [2,4,6,8,10]
y2 = [1,3,5,7,9]

plt.plot(x,y1,label = "Even")
plt.plot(x,y2,label = "odd")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

students = ["Om", "Raj", "Keya", "Neha"]
marks = [85, 90, 88, 70]

# Line Chart
plt.subplot(2, 2, 1)
plt.plot(students, marks, marker='o')
plt.title("Line Chart")

# Bar Chart
plt.subplot(2, 2, 2)
plt.bar(students, marks)
plt.title("Bar Chart")

# Pie Chart
plt.subplot(2, 2, 3)
plt.pie(marks, labels=students, autopct="%0.1f%%")
plt.title("Pie Chart")

# Scatter Plot
plt.subplot(2, 2, 4)
plt.scatter(students, marks)
plt.title("Scatter Plot")

plt.tight_layout()
plt.grid()
plt.show()
```
