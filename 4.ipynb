```python
import pandas as pd
```

```python
#loading the data set and hand made code 

data = {
    'Name' : ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age' : [24, 27, 22, 32, 29 ],
    'City' : ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'] ,
    'Marks': [85, 90, 78, 88, 92] 
}

df = pd.DataFrame(data)
print(df)
```

```python
print(df.isnull().sum())  # Check for missing values
```

```python
print(df.head(3))  # Display the first 3 rows
print(df.describe())  # Get summary statistics
print(df['Age'].mean())  # Calculate the mean age
print(df[df['Marks'] > 85])  # Filter rows where Marks > 85
print(df.sort_values(by='Marks', ascending=False))  # Sort by Marks in descending order
print(df.groupby('City')['Marks'].mean())  # Average Marks by City
```
