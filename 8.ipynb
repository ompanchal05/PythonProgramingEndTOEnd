```python
#Task B — Reverse a string without using reverse()
#Input: "Machine Learning"

input_str = "machine learning"
reversed_str = ""
reverse_index = len(input_str) - 1
while reverse_index >= 0:
    reversed_str += input_str[reverse_index]
    reverse_index -= 1
print(reversed_str)
```

```python
#Task C — Count frequency of each word in a sentence (use dictionary).
#Input: "data science is future and data is powerful"

input_sentence = "data science is future and data is powerful"
word_list = input_sentence.split()
word_frequency = {}
for word in word_list:
    if word in word_frequency:
        word_frequency[word] += 1
    else:
        word_frequency[word] = 1
print(word_frequency)
```
