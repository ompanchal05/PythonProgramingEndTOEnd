### Day 18 : Pandas: Data Cleaning, GroupBy & Aggregation

```python
import pandas as pd # Importing Pandas for Data manipulation and analysis
```

```python
data = {
    'Name': ['Om' , 'Raj','Keya' , 'Om' , 'Neha' , None] ,
    'Age' : [20 ,21,19,20,None ,22],
    'city' :['Ahmedabad', 'Mumbai' , None , 'Ahmedabad', 'Delhi', 'Mumbai'],
    'Marks' :[85,90,88,None , 70 ,85]
}
# importing the Data sets in the cell 
# this is the Data celll  And  I added the duplicates and None in the data to see the Groupby(), cleaning and aggregation in the panda 
```

```python
df = pd.DataFrame(data)
print (df)
# Step 1 : Printed the data to see how it's looked
```

```python
print(df.isnull().sum()) # Missing Value Joi apdee , column wise 
```

```python
df_drop = df.dropna()
print(df_drop) 
#Dropping missing the rows
```

```python
df_filled = df.copy()
df_filled['Age'].fillna(df['Age'].mean(),inplace = True )
df_filled['city'].fillna("Unknown", inplace=True)
df_filled['Marks'].fillna(df['Marks'].mean(), inplace =True)
df_filled['Name'].fillna("Unknown", inplace=True)

print(df_filled)

#Filling Misiing Values
```

```python
print(len(df_filled)) # Before Removing the duplicates 
df_clean = df_filled.drop_duplicates()
print(len(df_clean))
```

```python
df_clean['city'].replace("Mumbai" , "MUM" , inplace =True)
print(df_clean) # Replaceing the value 
print("\n")
df_clean['city'].replace("Ahmedabad", "Ahm" , inplace = True)
print(df_clean)
print("\n")
df_clean['city'].replace("Delhi" , "Del" , inplace = True)
print(df_clean)
print("\n")
```

```python
#Normalization the text , Uppercase
df_clean['Name'] = df_clean['Name'].str.upper()
print("\nAfter Converting Names to UPPERCASE:\n", df_clean)
```

```python
#Sorting the data 
df_sorted = df_clean.sort_values(by='Marks' , ascending=False)
print("\n Sorted By marks Desending order:\n" , df_sorted)
```

```python
print(df_clean.groupby("city")['Marks'].mean())
#AVG Marks By City 
#ke kai city ma ketla Students are clavers and all
```

```python
agg = df_clean.groupby("city")['Marks'].agg(['mean','min','max','count'])
print("\nAggregation Result:\n", agg)
```

```python
print("\n--- SUMMARY REPORT ---")
print("Total Students:", len(df_clean))
print("Average Age:", round(df_clean['Age'].mean(), 2))
print("Highest Marks:", df_clean['Marks'].max())
print("Top Student:\n", df_clean[df_clean['Marks'] == df_clean['Marks'].max()])
```
